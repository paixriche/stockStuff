```python
# Return the ticker of a company by using google search api
def tickerOfCompany(companyName):
    result = search("Yahoo Finance " + companyName)
    result = result[0].lower()
    return result.split('/')[-2]
```

```python
print(tickerOfCompany('Microsoft'))
```

```python
def stockHistory(companyName, period='max'):
    ticker = tickerOfCompany(companyName)
    company = yf.Ticker(ticker)
    
    df = company.history(period)
    
    stockPriceStart = 0
    stockPriceEnd = 0
    
    currentPrice = company.info['currentPrice']
    
    return df
```

```python
df = stockHistory('Microsoft')

plt.figure()
plt.plot(df['Close'])
plt.show()
```
